## Rescaling images

- The **rescale()** operation resizes an image by a given scaling factor.


- **rescale()** uses interpolation.


- The scaling factor can either be a single floating point value, or multiple values - one along each axis.


```python
A = np.arange(20).reshape(4, 5)
```

```python
print(A)
print(A.shape)
```

```python
%matplotlib inline

plt.imshow(A)
```

```python
B = rescale(A, scale = 2)
```

```python
print(B.shape)
```

```python
plt.imshow(B)
```

```python
print(B)
```

```python
C = rescale(A, scale = 2, preserve_range = True)

print(C)
```

```python
plt.imshow(C)
```

```python
print(np.max(A), np.max(B), np.max(C))
```

### Example: Satellite image

```python
image = io.imread("Satellite-No-Radar_20220621T20220621_1115.jpeg")

print(type(image))


print(np.max(image), np.min(image))
```

```python
image = io.imread("Satellite-No-Radar_20220621T20220621_1115.jpeg", as_gray= True)

print(type(image))


print(np.max(image), np.min(image))
```

```python
print(image.shape)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.imshow(image)
plt.show()
```

#### Rescaling the image

```python
rescaled_image = rescale(image, scale=1.5, preserve_range = True, anti_aliasing = True)
```

```python
print(np.max(image), np.max(rescaled_image))
```

```python
print(rescaled_image.shape)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.imshow(rescaled_image)
plt.show()
```
