# Derivatives in 2D

```python
import pyvista as pv

import numpy as np
import matplotlib.pyplot as plt
```

```python
mesh = pv.read("../unit-2/TURB_DRIVE_SUP/data.0010.vtk")
```

```python
print(mesh)
```

```python
rho = pv.get_array(mesh, "rho", preference = 'cell')
```

```python
rho_2D = rho.reshape(mesh.dimensions[0] - 1, mesh.dimensions[1] - 1)
```

```python
print(rho_2D.shape)
```

```python
x = np.linspace(mesh.bounds[0], mesh.bounds[1], mesh.dimensions[0] - 1)
y = np.linspace(mesh.bounds[2], mesh.bounds[3], mesh.dimensions[1] - 1)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
plt.figure(figsize = (13,10))

Z1 = plt.pcolor(X, Y, rho_2D, cmap = 'viridis', shading = 'auto')

CB1 = plt.colorbar(Z1)

plt.show()
```
