## Example 1:

Calculate the Fourier expansion of:
$f(x) = x^2$

```python
import sympy
import numpy as np

from sympy import fourier_series

from sympy import *
from sympy.abc import x
```

```python
def function(x):
    return x**2
```

```python


x = symbols('x')

function1 = x**2

result = fourier_series(function1, (x, -np.pi, np.pi))
```

## Example 2:

Calculate the Fourier expansion of:

$f(x) = 
\left\{
    \begin{array}{lr}
        0, & -\pi \leq x \leq -\frac{\pi}{2}\\
        1, & -\frac{\pi}{2} \leq x \leq +\frac{\pi}{2}\\
        0, & \frac{\pi}{2} \leq x \leq \pi
    \end{array}
\right\}$

```python
def function(x):
    
    if(-np.pi <= x <= -0.5*np.pi):
        return 0

    elif(-0.5*np.pi <= x <= +0.5*np.pi):
        return 1
    
    else:
        return 0
```

```python
x = np.arange(-np.pi, np.pi, 0.001)
```

```python
y = []

for i in range(len(x)):
    y.append(function(x[i]))
    
y = np.array(y)
```

```python
print(x.size, y.size)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))

plt.plot(x, y)

plt.show()
```

```python
def function_Fourier(x, n):
    f = np.zeros(x.size)
    f = f + 0.5
    for i in range(1,n+1):
        #print(i)
        y = (2/np.pi)*((-1)**(i-1))*(1/(2*i-1))*np.cos((2*i - 1)*x)
        f += y 
    return f
```

```python
plt.figure(figsize=(10, 5))

plt.plot(x, y, c ='blue')
plt.plot(x, function_Fourier(x, 1), c = 'red')
plt.plot(x, function_Fourier(x, 2), c = 'red', linestyle =':')
plt.plot(x, function_Fourier(x, 50), c = 'red', linestyle ='-.')

plt.show()
```
